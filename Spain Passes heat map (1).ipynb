```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mplsoccer
from statsbombpy import sb
```

```python
competitions=sb.competitions()
pd.set_option("display.max.rows",None)
competitions
```

```python
matches=sb.matches(competition_id=55, season_id=282)
matches
```

```python
events=sb.events(match_id=3942752)
events.columns
```

```python
#our team
team = "Spain"
#get list of games by our team, either home or away
match_ids = matches.loc[(matches["home_team"] == team) | (matches["away_team"] == team)]["match_id"].tolist()
#calculate number of games
no_games = len(match_ids)
no_games
```

```python
danger_passes = pd.DataFrame()

# Step 2: Loop through each match and retrieve event data
for _, match in matches.iterrows():
    match_id = match['match_id']
    events = sb.events(match_id=match_id)
    
# Optional: Check if events contain multiple DataFrames
    if isinstance(events, list):
        main_events = events[0]  # Access the first DataFrame if events is a list
    else:
        main_events = events  # If events is already a DataFrame, use it directly

    # Step 3: Process event data
    for period in [1, 2]:
        # Example: Filtering passes and shots, similar to the original logic
        team = "Spain" 
        mask_pass = (main_events['team'] == team) & (main_events['type'] == "Pass") & (main_events['pass_outcome'].isnull()) & (main_events['period'] == period)
        passes = main_events.loc[mask_pass, ["location", "pass_end_location","minute", "second", "player"]]

        mask_shot = (main_events['team'] == team) & (main_events['type'] == "Shot") & (main_events['period'] == period)
        shots = main_events.loc[mask_shot, ["minute", "second"]]

        shot_times = shots['minute'] * 60 + shots['second']
        shot_window = 15
        shot_start = shot_times - shot_window
        shot_start = shot_start.apply(lambda i: i if i > 0 else (period - 1) * 45)

        pass_times = passes['minute'] * 60 + passes['second']
        pass_to_shot = pass_times.apply(lambda x: True in ((shot_start < x) & (x < shot_times)).unique())

        danger_passes_period = passes.loc[pass_to_shot]
        danger_passes = pd.concat([danger_passes, danger_passes_period], ignore_index=True)
```

```python
danger_passes.head(5)
```

```python
num_danger_passes = len(danger_passes)
num_danger_passes
```

```python
# Assuming danger_passes DataFrame contains columns 'location', 'minute', 'second', 'player'
# And 'location' is a list of [x, y] coordinates, you can split these into separate columns for easier plotting
danger_passes[['start_x', 'start_y']] = pd.DataFrame(danger_passes['location'].tolist(), index=danger_passes.index)
# Extract 'end_x' and 'end_y' from 'end_location'
danger_passes[['end_x', 'end_y']] = pd.DataFrame(danger_passes['pass_end_location'].tolist(), index=danger_passes.index)

# Plotting
from mplsoccer import Pitch
pitch = Pitch(pitch_type='statsbomb', pitch_color='beige',line_color='black', goal_type='box')
fig,ax = pitch.draw(figsize=(14,8))

# Plot each pass
for i, row in danger_passes.iterrows():
    pitch.scatter(danger_passes['start_x'], danger_passes['start_y'], ax=ax, color='red', edgecolors='black', s=100, label='Start')
#uncomment it to plot arrows
    #pitch.arrows(row['start_x'], row['start_y'], row['end_x'], row['end_y'], width=2, headwidth=3, color='blue', ax=ax)
#add title
fig.suptitle('Location of danger passes by ' + team, fontsize = 30)
plt.show()
```

```python
pitch = Pitch(pitch_type='statsbomb', pitch_color='white',line_color='black', goal_type='box',line_zorder=2)
fig,ax = pitch.draw(figsize=(14,8))

# Calculate 2D histogram for pass start locations
bin_statistic = pitch.bin_statistic(danger_passes['start_x'], danger_passes['start_y'], statistic='count', bins=(6, 5), normalize=False)

# Normalize by the number of games
bin_statistic["statistic"] = bin_statistic["statistic"] / no_games

# Plot the heatmap
pcm = pitch.heatmap(bin_statistic, cmap='Reds', edgecolor='grey', ax=ax)

# Add a colorbar
fig.colorbar(pcm, ax=ax, location='right', fraction=0.02, pad=0.1)

# Set the title
fig.suptitle('Danger passes by ' + team + ' per game', fontsize=30)

# Show the plot
plt.show()
```

```python
pitch = Pitch(line_zorder=2, line_color='black')
fig, ax = pitch.grid(grid_height=0.9, title_height=0.06, axis=False,
                     endnote_height=0.04, title_space=0, endnote_space=0)
#get the 2D histogram
bin_statistic = pitch.bin_statistic(danger_passes.x, danger_passes.y, statistic='count', bins=(6, 5), normalize=False)
#normalize by number of games
bin_statistic["statistic"] = bin_statistic["statistic"]/no_games
#make a heatmap
pcm  = pitch.heatmap(bin_statistic, cmap='Reds', edgecolor='grey', ax=ax['pitch'])
#legend to our plot
ax_cbar = fig.add_axes((1, 0.093, 0.03, 0.786))
cbar = plt.colorbar(pcm, cax=ax_cbar)
fig.suptitle('Danger passes by ' + team + " per game", fontsize = 30)
plt.show()
```
