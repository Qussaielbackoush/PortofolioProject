```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mplsoccer
from statsbombpy import sb
```

```python
competitions=sb.competitions()
pd.set_option("display.max.rows",None)
competitions
```

```python
sb.matches(competition_id=55, season_id=282)
```

```python
events= sb.events(match_id=3943043)
events.head()
```

```python
events.columns
```

```python
all_shots = events[['team','type','player','location','shot_end_location','minute','shot_outcome','shot_type','shot_statsbomb_xg']]
all_shots.head(10)
```

```python
spain_shots = all_shots[(all_shots['team']=='Spain')& (all_shots['type']=='Shot')]
spain_shots.head(16)
```

```python
england_shots = all_shots[(all_shots['team']=='England')& (all_shots['type']=='Shot')]
england_shots.head(25)
```

```python
## Basic Visualization of the shots (arrows)
pitch = mplsoccer.Pitch(pitch_type='statsbomb', pitch_color='black',line_color='white', goal_type='box')
fig,ax = pitch.draw(figsize=(14,8))

for i, row in spain_shots.iterrows():
    start = row['location']
    end = row['shot_end_location']
    
    # Determine shot outcome color
    if row['shot_outcome'] in ['Goal', 'Saved']:
        color = 'green'  # Successful shot
    elif row['shot_outcome'] in ['Blocked', 'Off T']:
        color = 'red'  # Unsuccessful shot
    
    plt.arrow(start[0], start[1], end[0] - start[0], end[1] - start[1],
              head_width=1.5, head_length=1.5, fc=color, ec=color, alpha=0.7)
    fig.suptitle('Spain shots Vs England', fontsize=20,fontweight='bold',color='black')        
plt.show()
```

```python
nico_williams= spain_shots[(spain_shots['player']=='Nicholas Williams Arthuer')] 
n_p_williams=nico_williams.shape[0]
n_p_williams
```

```python
##Lets see Nico Williams Shots only on a half of a pitch
from mplsoccer.pitch import Pitch, VerticalPitch
pitch = mplsoccer.Pitch(pitch_type='statsbomb', positional=True, half=True, 
                        pitch_color='grass',line_color='white', goal_type='box')
fig,ax = pitch.draw(figsize=(14,8))

for i, row in nico_williams.iterrows():
    start = row['location']
    end = row['shot_end_location']
    
    # Determine pass outcome color
    if row['shot_outcome'] in ['Saved','Goal']:
        color = 'blue'  # Successful pass
    elif row ['shot_outcome'] in ['Blocked', 'Off T']:
        color = 'red'  # Unsuccessful pass (Out or Incomplete)
    
    
    # Plotting the start location of the shot
    plt.scatter(start[0], start[1],size,color, alpha=0.7)
    
    # Drawing line to the end location of the shot
    plt.plot([start[0], end[0]], [start[1], end[1]], color=color, alpha=0.7, linewidth=1.5)
    fig.suptitle("Nico Williams's Shots Vs England", fontsize=20,fontweight='bold',color='black')        
plt.show()
```
